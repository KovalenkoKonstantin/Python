Напиши программу, которая сгенерирует случайную игральную карту (2, 3, 4, 5, 6, 7, 8, 9, 10, валет, дама, король, туз)

одной из 4 мастей (червей, бубно, треф, пик)

```python
from random import choice

card = [2, 3, 4, 5, 6, 7, 8, 9, 10, 'валет', 'дама', 'король', 'туз']
color = ['червей', 'бубно', 'треф', 'пик']

print(choice(card), choice(color))
```
